```python
import google.generativeai as genai
```

```python
from google.colab import userdata
google_api_key=userdata.get('Api key')
```

```python
#configuration
genai.configure(api_key=google_api_key)
```

```python
#to view available gemini models
for model in genai.list_models():
  print(model.name)
```

```python
model=genai.GenerativeModel('gemini-2.5-flash')
```

```python
#response
response=model.generate_content("What is deep learning")
response.text
```

```python
#text wraping -make the text in an order

from IPython.display import Markdown
import textwrap
def demo(text):
  return Markdown(textwrap.indent(text,'>'))
```

```python
demo(response.text)
```

```python
response1=model.generate_content("What is generative ai")
response1.text
```

```python
demo(response1.text)
```
