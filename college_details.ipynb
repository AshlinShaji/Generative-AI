```python
import google.generativeai as genai

# Configure Gemini API
genai.configure(api_key='')

model = genai.GenerativeModel('gemini-2.5-flash')

# College Information (Context)
college_info = """
College Name: ABC Institute of Technology
Location: Kottayam, Kerala
Courses Offered:
- B.Tech (CSE, ECE, EEE, Mechanical)
- M.Tech
- MCA
- Diploma Programs

Admission:
- Based on entrance exam and merit
- Management quota available

Fees:
- B.Tech: ₹75,000 per year
- MCA: ₹60,000 per year

Facilities:
- Library
- Computer Labs
- Hostel
- Sports Ground
- Wi-Fi Campus

Placements:
- Top recruiters: TCS, Infosys, Wipro, IBM
- Average package: 4 LPA

Contact:
- Phone: 9876543210
- Email: info@abcit.edu
"""

print(" Welcome to ABC College AI Chatbot")
print("Ask about courses, fees, admission, placement, facilities, etc.")
print("Type 'exit' to quit\n")

while True:
    user_query = input("You: ")

    if user_query.lower() == "exit":
        print("Chatbot: Thank you for visiting ABC Institute of Technology!")
        break

    prompt = f"""
You are a college website chatbot.
Answer ONLY using the information below.
If the answer is not available, say "Please contact the college office for more details."

College Information:
{college_info}

User Question:
{user_query}
"""

    response = model.generate_content(prompt)
    print("Chatbot:", response.text)
```
